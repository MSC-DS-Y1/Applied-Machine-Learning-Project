### Mutual Information (Target)

Mutual information feature selection - for redundant features


Calculate the mutual information score on each of the 20 selected input variables in comparison with the target vector. The higher the input variable, the better related to the target value.

```python
#see if x is descrete or continuous
discrete_features = x_vars.dtypes == "int64" 
#discrete_features['Age_at_enrollment'] = False #change this manually to ensure age is seen as a continuous variable
#previously had set this as continuous, but it might be better if this is set as categorical as in the dataset definitions?

from sklearn.feature_selection import mutual_info_classif, mutual_info_regression

def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_classif(X, y, discrete_features=discrete_features, random_state=42) # assigning whether the x_variable is descrete or continuous, always keeping the random state consistent for the same results from one iteration to another
    mi_scores = pd.Series(mi_scores, name = "MI Scores", index = X.columns)
    mi_scores = mi_scores.sort_values(ascending = False)
    return mi_scores

mi_scores = make_mi_scores(x_vars, y_vars,discrete_features)

print(mi_scores)

def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores with Target")


plt.figure(dpi=100, figsize=(8, 5))
plot_mi_scores(mi_scores)
```

### Mutual Information (Pairwise)

For irrelevant features, need to calculate the pairwise mutual information . 
High Mutual information scores for pairs of variables means that they provide similar information to the dataset and some can be removed

```python
x_vars.shape[1]

def pairwise_mi(df, discrete_features):
    n = df.shape[1]
    mi_matrix = np.zeros((n,n))
    
    for i in range(n):
        for j in range(n):
            #start with equal pairs and mark is not applicable
            if i == j:
                mi_matrix[i,j] = np.nan
            else:
                x = df.iloc[:,i]
                y = df.iloc[:,j]
                
                #first case: if both variables are discrete
                if discrete_features[i] and discrete_features[j]:
                    mi = mutual_info_classif(x.values.reshape(-1,1), y.values, discrete_features = True, random_state=42)[0]
                #now if both variables are continuous
                elif not discrete_features[i] and not discrete_features[j]:
                    mi = mutual_info_regression(x.values.reshape(-1,1), y.values, discrete_features=False, random_state=42)[0]
                #now one is continuous and another is discrete- prefer to take a regression case than a classification case, leaving descrete_feature as auto in this case
                else:
                    mi = mutual_info_regression(x.values.reshape(-1,1), y.values, random_state=42)[0]
                
                mi_matrix[i,j] = mi
    return pd.DataFrame(mi_matrix,index = df.columns, columns = df.columns)
                


#generate pairwise MI matrix
mi_matrix = pairwise_mi(x_vars, discrete_features)
```

```python
#plotting the pairwise mi scores to visualise in a heatmap

plt.figure(figsize = (10,10))
plt.imshow(mi_matrix, cmap = "plasma", interpolation = "nearest", origin = "upper")
#plt.colorbar(label="Mutual Information")

labels = x_vars.columns

plt.xticks(ticks=np.arange(len(labels)), labels=labels, rotation=45, ha="right", fontsize=10)
plt.yticks(ticks=np.arange(len(labels)), labels=labels, fontsize=10)
plt.title("Pairwise Mutual Information Heatmap", fontsize=14)
plt.tight_layout()
plt.show()
```

A threshold for the Target MI scores can be assigned based on the values in the plot
Similarly, a threshold based on the MI matrix values can be assigned for the pairwise MI scores.

The final variables to be kept are those which first meet the MI Target threshold, and then the pairwise MI threshold.

```python
#setting the thresholds, these can be altered
mi_threshold_target = 0.02
mi_threshold_pairwise = 0.5


relevant_features = [col for col in mi_scores.index if mi_scores[col] > mi_threshold_target]
print(f"Relevant features based on target MI: {relevant_features}")

remaining_features = relevant_features.copy()


features_to_drop = []

# Check pairwise MI between remaining features
for i in range(len(remaining_features)):
    for j in range(i + 1, len(remaining_features)):
        # If pairwise MI is above the threshold, drop one feature
        mi_value = mi_matrix.loc[remaining_features[i], remaining_features[j]]
        if mi_value > mi_threshold_pairwise:
            features_to_drop.append(remaining_features[j])

final_features = [feature for feature in remaining_features if feature not in features_to_drop]
print(f"Final selected features: {final_features}")
```
