```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('uncleaned_dataset.csv', sep=';')
data.head()
```

```python
data.describe()
```

Renaming some variables and rounding up curricular units

```python
data = data.rename(columns=({'Nacionality':'Nationality'}))
data = data.rename(columns=({'Daytime/evening attendance	':'Daytime/evening attendance'}))
data['Curricular units 1st sem (grade)'] = data['Curricular units 1st sem (grade)'].round(2)
data['Curricular units 2nd sem (grade)'] = data['Curricular units 2nd sem (grade)'].round(2)
data['Target'] = data['Target'].map({'Dropout':0,'Enrolled':1,'Graduate':2})
```

```python
data.to_csv('cleaned_dataset.csv', index=False)
```

```python
#Spearman
#Encoding the data as .corr does not work with string variables, thus we convert them to 1, 2, ect.
encodedData = pd.get_dummies(data, drop_first=True)

# Calculate Pearson correlations for all variables
correlation_matrix = encodedData.corr(method='pearson')

# Display the correlation matrix
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
sns.set_theme(font_scale=0.6)
```

```python
from sklearn.feature_selection import mutual_info_classif, mutual_info_regression

encodedData = pd.get_dummies(data)

mi_matrix = np.zeros((encodedData.shape[1], encodedData.shape[1]))

#Calculate mutual information for all pairs of features
for i in range(encodedData.shape[1]):
    for j in range(i, encodedData.shape[1]):
        # If both columns are continuous, use mutual_info_regression
        if encodedData.iloc[:, i].dtype in ['float64', 'int64'] and encodedData.iloc[:, j].dtype in ['float64', 'int64']:
            mi_value = mutual_info_regression(encodedData.iloc[:, [i]], encodedData.iloc[:, j])[0]
        # If both columns are categorical, use mutual_info_classif
        elif encodedData.iloc[:, i].dtype == 'uint8' and encodedData.iloc[:, j].dtype == 'uint8':
            mi_value = mutual_info_classif(encodedData.iloc[:, [i]], encodedData.iloc[:, j])[0]
        else:
            mi_value = 0  # Assign 0 if not comparable
        
        mi_matrix[i, j] = mi_value
        mi_matrix[j, i] = mi_value  # Matrix is symmetric

#Convert the matrix to DataFrame for easier visualization
mi_matrix_df = pd.DataFrame(mi_matrix, columns=encodedData.columns, index=encodedData.columns)

sns.heatmap(mi_matrix_df, annot=False, cmap='coolwarm')
sns.set_theme(font_scale=0.6)
plt.title("Mutual Information Matrix")
plt.show()
```

```python
# List of columns that are numerical and suitable for boxplots
numerical_columns = [
    'Application order', 'Previous qualification (grade)', 'Admission grade',
    'Age at enrollment', 'Curricular units 1st sem (credited)', 'Curricular units 1st sem (enrolled)',
    'Curricular units 1st sem (evaluations)', 'Curricular units 1st sem (approved)', 
    'Curricular units 1st sem (grade)', 'Curricular units 1st sem (without evaluations)',
    'Curricular units 2nd sem (credited)', 'Curricular units 2nd sem (enrolled)',
    'Curricular units 2nd sem (evaluations)', 'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)', 'Curricular units 2nd sem (without evaluations)',
    'Unemployment rate', 'Inflation rate', 'GDP'
]

n_cols = 4
n_rows = (len(numerical_columns) + n_cols - 1) // n_cols

# Set the size of the plot
plt.figure(figsize=(n_cols * 4, n_rows * 4))

# Create a grid of boxplots
for i, column in enumerate(numerical_columns, start=1):
    plt.subplot(n_rows, n_cols, i)
    sns.boxplot(x=data[column])
    plt.title(f'Boxplot for {column}', fontsize=12)
    plt.xlabel(column, fontsize=10)
    plt.tight_layout()

plt.show()
```

```python
thresholds = [
    {"Name": "Marital status", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Application mode", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Application order", "UpperThreshold": 3.5, "LowerThreshold": 0},
    {"Name": "Course", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Daytime/evening attendance", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Previous qualification", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Previous qualification (grade)", "UpperThreshold": 165, "LowerThreshold": 100},
    {"Name": "Nationality", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Mother's qualification", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Father's qualification", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Mother's occupation", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Father's occupation", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Admission grade", "UpperThreshold": 157, "LowerThreshold": 0},
    {"Name": "Displaced", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Educational special needs", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Debtor", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Tuition fees up to date", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Gender", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Scholarship holder", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Age at enrollment", "UpperThreshold": 23.5, "LowerThreshold": 0},
    {"Name": "International", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Curricular units 1st sem (credited)", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Curricular units 1st sem (enrolled)", "UpperThreshold": 10, "LowerThreshold": 0},
    {"Name": "Curricular units 1st sem (evaluations)", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Curricular units 1st sem (approved)", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Curricular units 1st sem (grade)", "UpperThreshold": 0, "LowerThreshold": 0.1},
    {"Name": "Curricular units 1st sem (without evaluations)", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Curricular units 2nd sem (credited)", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Curricular units 2nd sem (enrolled)", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Curricular units 2nd sem (evaluations)", "UpperThreshold": 15, "LowerThreshold": 0},
    {"Name": "Curricular units 2nd sem (approved)", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Curricular units 2nd sem (grade)", "UpperThreshold": 0, "LowerThreshold": 0.1},
    {"Name": "Curricular units 2nd sem (without evaluations)", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Unemployment rate", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Inflation rate", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "GDP", "UpperThreshold": 0, "LowerThreshold": 0},
    {"Name": "Target", "UpperThreshold": 0, "LowerThreshold": 0}
]

def remove_outliers(data, thresholds):
    # Iterate over each feature in the thresholds list
    for threshold in thresholds:
        column = threshold['Name']
        upper_threshold = threshold['UpperThreshold']
        lower_threshold = threshold['LowerThreshold']

        # If thresholds are non-zero, apply filtering to remove outliers
        if upper_threshold != 0:
            data = data[(data[column] <= upper_threshold)]
            
        if lower_threshold != 0:
            data = data[(data[column] >= lower_threshold)]
    
    return data

cleaned_data = remove_outliers(data, thresholds)



n_cols = 4
n_rows = (len(numerical_columns) + n_cols - 1) // n_cols

# Set the size of the plot
plt.figure(figsize=(n_cols * 4, n_rows * 4))

# Create a grid of boxplots
for i, column in enumerate(numerical_columns, start=1):
    plt.subplot(n_rows, n_cols, i)
    sns.boxplot(x=cleaned_data[column])
    plt.title(f'Boxplot for {column}', fontsize=12)
    plt.xlabel(column, fontsize=10)
    plt.tight_layout()

plt.show()

print(cleaned_data.shape)
```
