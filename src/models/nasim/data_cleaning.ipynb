```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 


data = pd.read_csv('uncleaned_dataset.csv', sep=';')
data.head()
```

```python
data.info()
```

```python
# Converting floats to ints 
# m = data.select_dtypes(np.number)
# data[m.columns]= m.round().astype('int64')

# Fixing column names
data = data.rename(columns={"Nacionality":"Nationality",
                            "Daytime/evening attendance\t":"Daytime/evening attendance"})

data = data.drop(['Previous qualification (grade)', 'Admission grade'], axis='columns')

data.info()
```

```python
y = data['Target']
x = data.drop(['Target'], axis=1)
```

### Dataset Imbalance

```python
count_class = y.value_counts() # Count the occurrences of each class
plt.bar(count_class.index, count_class.values)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution')
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

x, y = SMOTE().fit_resample(x, y)

sorted(Counter(y).items()) # resulting balanced dateframe
```

### Spearman Correlation Heatmap

```python
# Pearson Correlation
corr = data.select_dtypes(['int64']).corr(method='pearson')

# plot the heatmap
sns.heatmap(corr, cmap='coolwarm')

sns.set_theme(font_scale=0.4)
```

### Target MI

```python
from sklearn.feature_selection import mutual_info_classif

mi = mutual_info_classif(x, y, random_state=42)
mi
```

```python
# Create a DataFrame to store feature names and their mutual information values
mi_df = pd.DataFrame({'Feature': x.columns, 'Mutual Information': mi})

# Sort by mutual information values
mi_df = mi_df.sort_values(by='Mutual Information', ascending=False)

sns.barplot(x='Mutual Information', y='Feature', data=mi_df)
plt.title('Mutual Information between Features and Target')
plt.xlabel('Mutual Information')
plt.ylabel('Features') 
```

```python
discrete = x.dtypes == "int64" 

mi = mutual_info_classif(x, y, discrete_features=discrete, random_state=42)
mi
```

```python
# Create a DataFrame to store feature names and their mutual information values
mi_df = pd.DataFrame({'Feature': x.columns, 'Mutual Information': mi})

# Sort by mutual information values
mi_df = mi_df.sort_values(by='Mutual Information', ascending=False)

sns.barplot(x='Mutual Information', y='Feature', data=mi_df)
plt.title('Mutual Information between Features and Target')
plt.xlabel('Mutual Information')
plt.ylabel('Features') 
```

#### Data Normalisation

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
```

```python
mi_scores = pd.Series(mi, name = "MI Scores", index = x.columns) # create PD series with MI scores
mi_scores = mi_scores.sort_values(ascending = False)

mi_scores = mi_scores.to_frame() # converts from pandas object to df
mi_scores_scaled = min_max_scaler.fit_transform(mi_scores) # scales MI scores

values = mi_scores_scaled.flatten()
indices = mi_scores.index

mi_scores = pd.DataFrame(values, index=indices, columns=["Values"])
mi_scores
```

```python
# Plot the bar chart
sns.barplot(y=mi_scores.index, x='Values', data=mi_scores)

# Add labels and title
plt.title('Mutual Information between Features and Target (Normalised Values)')
plt.xlabel('Mutual Information')
plt.ylabel('Features') 

# Show the plot
plt.show()
```

#### MI by Clustering

```python
data_demo = x[['Marital status',
                  'Nationality',
                  'Displaced',
                  'Gender',
                  'Age at enrollment',
                  'International']]

data_socio_eco = x[['Mother\'s qualification',
                       'Father\'s qualification',
                       'Mother\'s occupation',
                       'Father\'s occupation',
                       'Educational special needs',
                       'Debtor',
                       'Tuition fees up to date',
                       'Scholarship holder']]

data_macro_eco = x[['Unemployment rate',
                       'GDP',
                       'Inflation rate']]

data_enrollment = x[['Application mode',
                        'Application order',
                        'Course',
                        'Daytime/evening attendance',
                        'Previous qualification']]

data_first_sem = x[['Curricular units 1st sem (credited)',
                       'Curricular units 1st sem (enrolled)',
                       'Curricular units 1st sem (evaluations)',
                       'Curricular units 1st sem (approved)',
                       'Curricular units 1st sem (grade)',
                       'Curricular units 1st sem (without evaluations)']]

data_second_sem = x[['Curricular units 2nd sem (credited)',
                       'Curricular units 2nd sem (enrolled)',
                       'Curricular units 2nd sem (evaluations)',
                       'Curricular units 2nd sem (approved)',
                       'Curricular units 2nd sem (grade)',
                       'Curricular units 2nd sem (without evaluations)']]

clusters = [data_demo, data_socio_eco, data_macro_eco, data_enrollment, data_first_sem, data_second_sem]
```

```python
def mi_score(cluster, target): 
    discrete = cluster.dtypes == "int64" 
    mi = mutual_info_classif(cluster, target, discrete_features=discrete, random_state=42)

    mi_scores = pd.Series(mi, name = "MI Scores", index = cluster.columns) # create PD series with MI scores
    mi_scores = mi_scores.sort_values(ascending = False)

    mi_df = mi_scores.to_frame()  # Convert the Series to a DataFrame
    return mi_df
```

```python
# Create plot
fig, axes = plt.subplots(3, 2, figsize=(12, 6), sharey=False)
axes = axes.flatten()  # Flatten the 2D array to make indexing easier
n = 0

for cluster in clusters:
    mi_plot = mi_score(cluster, y)
   
    sns.barplot(y=mi_plot.index, x=mi_plot['MI Scores'], data=mi_plot, ax=axes[n])

    # Add labels and title
    axes[n].set_title(f'Mutual Information between Cluster and Target (Non-Normalised Values)')
    axes[n].set_xlabel('Mutual Information')
    axes[n].set_ylabel('Features') 
    
    n += 1 # axis counter
    
plt.tight_layout()
plt.show()    
```

#### MI by Clustering (Normalised)

```python
# Create plot
fig, axes = plt.subplots(3, 2, figsize=(12, 6), sharey=False)
axes = axes.flatten()  # Flatten the 2D array
n = 0

for cluster in clusters:
    mi_plot = mi_score(cluster, y)
    
    scaled = min_max_scaler.fit_transform(mi_plot) # scales MI scores
    
    mi_scores_scaled = pd.DataFrame(data = scaled.flatten(), index = mi_plot.index, columns=['Values'])
   
    sns.barplot(y=mi_scores_scaled.index, x=mi_scores_scaled['Values'], data=mi_scores_scaled, ax=axes[n])

    # Add labels and title
    axes[n].set_title(f'Mutual Information between Cluster and Target (Normalised Values)')
    axes[n].set_xlabel('Mutual Information')
    axes[n].set_ylabel('Features') 
    
    n += 1 # axis counter
    
plt.tight_layout()
plt.show()    
```
